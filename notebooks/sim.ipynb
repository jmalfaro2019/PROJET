```python
import matplotlib.pyplot as plt
import sys
sys.path.append('..') 
from src.simulation import run_monte_carlo

# 1. Correr la simulacin (Clculo)
history = run_monte_carlo(n_initial_neutrons=1000, n_generations_max=500)

# 2. Visualizar los datos (Grfico)
plt.figure(figsize=(10, 6))
plt.plot(history, color='blue', linewidth=2, label='Neutron Population')
plt.axhline(y=history[0], color='red', linestyle='--', alpha=0.5, label='Start')

plt.title("Monte Carlo Simulation Results")
plt.xlabel("Generations")
plt.ylabel("Number of Neutrons")
plt.grid(True, alpha=0.3)
plt.legend()

plt.show() # <--- Esta es la clave para que aparezca!
```
